```python
import pandas as pd
```

```python
data = pd.read_csv('../data/Apartments.csv')
data.info()
```

```python
data.sample(10)
```

```python
data.dropna(inplace=True)
data.info()
```

```python
data['total_rooms'] = data['Bathrooms'] + data['Bedrooms']
data = data.drop(['Bathrooms','Bedrooms'],axis=1)
```

```python
data['PrivateGarden'] = data['PrivateGarden'].map({'Yes':1,'No':0})
```

```python
data['price'] = data['price'].replace(r'[^\d]','',regex = True).astype(float)
```

```python
data.info()
```

```python
data = pd.get_dummies(data)
data.sample(5)
```

We can delete one dummy variable from each feature so this variable will be the baseline and will be represented by the intercept in our regression model

```python
data = data.drop(['Payment_Cash','Ownership_Resale', 'Status_Ready'], axis=1)
```

```python
data.to_csv('../data/ProcessedData.csv', index = False)
```
